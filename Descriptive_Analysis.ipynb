```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from ggplot import mtcars
```

```python
#Get the header data for 'mtcars' and store it in variable 'header_data'. Display the header data using print() command.
header_data = mtcars.head()
header_data
```

```python
mtcars.describe()
```

```python
#Calculate the mean of each of the columns of mtcars
mtcars.mean()
```

```python
#Get the mean of each row (car) and store it in variable 'mtcarsMeanCarWise'
mtcarsMeanCarWise = mtcars.mean(axis=1)
mtcarsMeanCarWise
```

```python
#Get the median of each column and store it in variable 'medianCarsFeatures'
medianCarsFeatures = mtcars.median()
medianCarsFeatures
```

```python
#Plot the histograms for each of the columns in mtcars using hist() function with bin-size as 10.
mtcars.hist(bins=10,figsize=(9,10))
```

```python
#Plot the density curves for each column of mtcars using 'for' loop and plot command
for cols in mtcars.columns:
    if (cols != 'name'):
        mtcars[cols].plot(kind="density",figsize=(4,2))
        plt.ylabel("")
        plt.title(cols)
        plt.show()
```

```python
#Plot the correlation for the following attributes: mpg, disp, hp, wt, qsec. You can use the scatter_matrix command present in pandas.plotting library
corr_cols = ['mpg', 'disp', 'hp', 'wt', 'qsec']
pd.plotting.scatter_matrix(mtcars[corr_cols],figsize=(11,9),alpha = 0.8,marker = "*",grid= True)
```

```python
#Use describe and quantile to calculate the Inter-Quantile spread of 'mpg' attribute of mtcars
print(mtcars['mpg'].quantile())
mtcars['mpg'].describe()
```

```python
#Plot the box plot for 'mpg' column of mtcars
plt.boxplot(mtcars['mpg'],notch = True)
```

```python
#Calculate the mean, standard deviation and variance of the 'mpg' column
print(mtcars['mpg'].mean())
print(mtcars['mpg'].var())
print(mtcars['mpg'].std())
```

```python
#Extract cars with automatic transmission and cars with manual transmission into separate variables 'mtCarsAutomatic' and 'mtCarsManual' respectively
mtCarsAutomatic = mtcars.loc[mtcars.am == 1]
mtCarsManual = mtcars[mtcars.am == 0]
```

```python
#Plot the box plot for 'mpg' (miles per gallons) for each of the subsets of cars grouped by transmission (automatic/ manual)
pd.DataFrame(mtcars[['mpg','am']]).boxplot(by='am',figsize=(5,5))
```

```python
#Calculate the skeweness of each column of 'mtcars'
mtcars.skew().sort_values
```

```python
#Generate 10000 random normally distributed numbers and plot the density curve along with the median and mean
norm_data = pd.DataFrame(np.random.normal(size=100000))
norm_data.plot(kind="density")
```

```python
#Generate 10000 exponentially distributed numbers using np.random.exponential(size=100000) and plot the density curve along with the median and mean.
skewed_data = pd.DataFrame(np.random.exponential(size=100000))

skewed_data.plot(kind="density",figsize=(10,10),xlim=(-1,5))
```
